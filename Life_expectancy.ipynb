```python
life.dropna(axis=0, inplace=False)

life.isnull().sum()
```

```python
life.isnull().sum()
```

```python
# replace with mean, mode or median
mean_val = life["Population"].mean()
median_val = life["Population"].median()
mode_val = life["Population"].mode()

life.fillna(value=mode_val, inplace=True)

life.head(50)
```

```python
life.head(50)
```

```python
from sklearn.impute import SimpleImputer

expenditure = life["Total expenditure"]

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer = imputer.fit(expenditure)

expenditure = imputer.transform(expenditure)

```

```python
life.fillna(axis=0, method="ffill", inplace=True)

life.fillna(axis=0, method="bfill", inplace=True)

life.isnull().any()
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean', missing_values=np.nan)
imputer = imputer.fit(life[['Population']])

life[['Population']] = imputer.transform(life[['Population']])
life.head()
```
